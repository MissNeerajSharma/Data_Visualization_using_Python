## 3. Remove duplicate rows 


```python
#Remove duplicate rows
df2=df.drop_duplicates()
```

## 4. Remove rows which have negative values in Quantity column 

```python
df1 = df[df.Quantity > 0]
```

```python
df1.describe().round(2)
```

## 5. Add the columns - Month, Day and Hour for the invoice 


```python
import datetime as dt
df1['year'] = pd.DatetimeIndex(df1['InvoiceDate']).year

```

```python
df1['year'].head(2)
```

```python
df1['Month'] = pd.DatetimeIndex(df1['InvoiceDate']).month
```

```python
df1['Day'] = pd.DatetimeIndex(df1['InvoiceDate']).day
df1['Hour'] = pd.DatetimeIndex(df1['InvoiceDate']).hour
df1['Minutes'] = pd.DatetimeIndex(df1['InvoiceDate']).minute
```

```python
df1.head()
```

## 6. How many orders made by the customers? 


```python
index = pd.Index(df1['Quantity'])
print(f"Total orders made by the customers = {index.value_counts().sum()}")
```

## 7. TOP 5 customers with higher number of orders 

```python
temp=df1.groupby(by=['CustomerID'], as_index=False)['InvoiceNo'].count().head(5)
tempsort=temp.sort_values(by='InvoiceNo', ascending=False).head()
print(tempsort)
```

## 8.How much money spent by the customers?

```python
df1['Revenue'] = df1['Quantity'] * df1['UnitPrice']
df1['Revenue'].head(2)
```

```python
money_spent = df1.groupby(by=['CustomerID','Country'], as_index=False)['Revenue'].sum()
print(money_spent)
```

### 9. TOP 5 customers with highest money spent 


```python

money_spent.sort_values(by='Revenue', ascending=False).head(5)
```

### 10. How many orders per month? 


```python
plt.figure(figsize=(15, 5))
ax = df1.groupby('InvoiceNo')['Month'].unique().value_counts().sort_index()
ax.plot.bar(color="green",width=0.8)
plt.ylabel('Total number of orders')
plt.xlabel('Months Number')
plt.title('Numbers of order per month')
```

### 11. How many orders per day? 


```python
plt.figure(figsize=(15, 5))
ax2 = df1.groupby('InvoiceNo')['Day'].unique().value_counts().sort_index()
ax2.plot.bar(color="green",width=0.8)
plt.ylabel('No of orders')
plt.xlabel('Date number of months')
plt.title('Numbers of order per day')
```

### 12. How many orders per hour?

```python
df1['Hour'].value_counts()
```

```python
ax = df1.groupby('InvoiceNo')['Hour'].unique().value_counts().iloc[:-1].sort_index().plot(kind='bar',color='green',figsize=(15,6))
ax.set_xlabel('Hour',fontsize=13)
ax.set_ylabel('Number of Orders',fontsize=13)
ax.set_title('Number of Hours',fontsize=13)
ax.set_xticklabels(range(6,21), rotation='horizontal', fontsize=13)
plt.show()
```

### 13. How many orders for each country? 


```python
orders = df1.groupby('Country')['InvoiceNo'].count().sort_values(ascending=False)
orders
```

### 14. Orders trend across months 


```python
# trend plot
fig = plt.figure(figsize=(15, 7))
x1 = fig.add_subplot(111)

x1.set_xlabel('Month',fontsize=15)
x1.set_ylabel('Total Amount Spent ($)',fontsize=15)
x1.set_title('Total Amount Spent for Months',fontsize=15)

x1.bar(df1.groupby(by=['Month'])['Revenue'].sum().index.tolist()[1:],
        df1.groupby(by=['Month'])['Revenue'].sum()[1:],
        alpha=0.85,
        label='Amount Spent by Month',color='green')

x2 = x1.twinx()# tread line

x2.set_ylabel('Percentage Change of Total Amount Spent (%)',fontsize=15)

#pct_change()Percentage change between the current and a prior element.
x2.plot(df1.groupby(by=['Month'])['Revenue'].sum().index.tolist()[1:], 
         df1.groupby(by=['Month'])['Revenue'].sum()[1:].pct_change().fillna(0)*100,
         label='Percentage change of total amount spent (%)',
         color='red')

x1.legend(loc='upper left')
x2.legend(loc='upper right')

fig.tight_layout()
plt.show()
```

### 15. How much money spent by each country

```python
group_country_amount_spent = df1.groupby('Country')['Revenue'].sum().sort_values().head(10)

```

```python
plt.subplots(figsize=(15,8))
group_country_amount_spent.plot(kind='barh', fontsize=12, color='green')
plt.xlabel('Money Spent (Dollar)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Money Spent by different Countries', fontsize=12)
plt.show()
```
