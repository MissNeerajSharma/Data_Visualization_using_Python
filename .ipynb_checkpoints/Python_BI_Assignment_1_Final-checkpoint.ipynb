# <font color='blue'> <center>Part 2. Data clean up – correcting the data types</center>

### <font color='blue'> Que)a. Which all variables need to be brought to numeric types?

```python
print(f"{df1.dtypes}\n \nFrom above we can say columns Reviews,Installs,Price needs to brought into numeric datatype")
```

```python
df1['Reviews'] = df1['Reviews'].astype(int)
df1['Size'] = df1['Size'].astype(int)
#df1['Installs']=df1['Installs'].astype(int)

# Additional conversion
df1['Last Updated'] = df1['Last Updated'].astype('datetime64[ns]')
```

```python
print(f"Change Datatype Checking..\n\n {df1.dtypes}")
```

### <font color='blue'> Que)b. Price variable – remove $ sign and convert to float


```python
df1['Price']  = df1['Price'] .replace({'\$':''}, regex = True)
```

```python
df1.drop(labels=df1[df1['Price']=='Everyone'].index, inplace = True)

df1['Price']= df1['Price'].astype('float')
```

```python
df1.Price.value_counts()
```

###     Finally !!! $ symbol removed from Price columns

## <center><font color="blue">Part 11. Content rating vs. size vs. rating – 3 variables at a time</center>

###  <font color="blue">11.a)Create 5 buckets (20% records in each) based on Size

```python
sns.distplot(df1["Size"], bins=5)
plt.title("columnname -->Size Histrogram")
plt.show()
```

### <font color="blue">11.b). By Content Rating vs. Size buckets, get the rating (20th percentile) for each combination

```python
bins=[0, 20000, 40000, 60000, 80000, 100000]
df1['Bucket Size'] = pd.cut(df1['Size'], bins, labels=['0-20k','20k-40k','40k-60k','60k-80k','80k-100k'])
pd.pivot_table(df1, values='Rating', index='Bucket Size', columns='Content Rating')
```

```python
df1.Size.quantile([0.2, 0.4,0.6,0.8])
```

### <font color="blue"> 11.c). Make a heatmap of this
* i. Annotated
* ii. Greens color map

```python
temp3=pd.pivot_table(df1, values='Rating', index='Bucket Size', columns='Content Rating', aggfunc=lambda x:np.quantile(x,0.2))
temp3
```

```python
f,ax = plt.subplots(figsize=(5, 5))
sns.heatmap(temp3, annot=True, linewidths=.5, cmap='Greens',fmt='.1f',ax=ax)
plt.title("Heat Map for finding correlation with App size vs Content Rating")
plt.show()
```

#### <font color="blue"> 11.d). What’s your inference? Are lighter apps preferred in all categories? Heavier? Some

#### Based on above visualization , We can conclude that ,No Its not true that lighter apps are preferred in all categories.
#### Because apps with size 40k-60k and 80k-100k have got the highest rating in all cateegories. 
#### So, in general we can conclude that heavier apps are preferred in all categories.
